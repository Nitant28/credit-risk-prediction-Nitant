```python
import os
import time

# Path to your main project folder (adjust if needed)
project_path = "."

# Current time (now)
current_time = time.time()

for root, dirs, files in os.walk(project_path):
    for name in files:
        file_path = os.path.join(root, name)
        # set modification and access time to NOW
        os.utime(file_path, (current_time, current_time))

print("All file timestamps updated to today!")
```

```python
import os
import time

# Path to your project folder (current directory)
project_path = "."

# Current time
current_time = time.time()

# Update timestamps for folders
for root, dirs, files in os.walk(project_path):
    for d in dirs:
        folder_path = os.path.join(root, d)
        os.utime(folder_path, (current_time, current_time))

print("All folder timestamps updated!")
```
